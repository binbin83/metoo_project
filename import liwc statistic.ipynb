```python
########### we set a proper index
def set_index_liwc(df):
    find_nb = re.compile(r'(?<=tweet)(.*)(?=_)')
    I=[]
    for i in range(len(df)):
        text=df['Filename'].iloc[i]
        k=int(find_nb.findall(text)[0])
        I.append(k)
    df['num_tweet']=I
    df.set_index('num_tweet',inplace=True)
############## we add the interesting colum to the initial data frame for the english version
def add_liwc_en(df_liwc,df):
    df.reset_index(inplace=True,drop=True)
    df['dic']=df_liwc['Dic']# the percentage of word analyse by the LIWC dictionaries
    df['sixltr']=df_liwc['Sixltr']#the percentage of word with more than 6 letter
    #pronums
    df['i']=df_liwc['i']
    df['we']=df_liwc['we']
    df['shehe']=df_liwc['shehe']
    df['they']=df_liwc['they']
    df['verb']=df_liwc['verb']
    #affect
    df['affect']=df_liwc['affect']
    df['posemo']=df_liwc['posemo']
    df['negemo']=df_liwc['negemo']
    #negemo
    df['anxiety']=df_liwc['anx']
    df['anger']=df_liwc['anger']
    df['sad']=df_liwc['sad']
    #attitude
    df['social']=df_liwc['social']
    df['cogproc']=df_liwc['cogmech']
    df['certain']=df_liwc['certain']
    df['past']=df_liwc['past']
    df['present']=df_liwc['present']
    df['future']=df_liwc['future']
    #other
    df['work']=df_liwc['work']
    df['home']=df_liwc['home']
    df['health']=df_liwc['health']
    df['death']=df_liwc['death']
    df['family']=df_liwc['family']
    df['time']=df_liwc['time']
    df['body']=df_liwc['body']
    df['incl']=df_liwc['incl']
    
######################################################## now we do the same for the french one, keeping the same name of colum
def add_liwc_fr(df_liwc,df):
    df.reset_index(inplace=True,drop=True)
    df['dic']=df_liwc['Dic']# the percentage of word analyse by the LIWC dictionaries
    df['sixltr']=df_liwc['Sixltr']#the percentage of word with more than 6 letter
    #pronums
    df['i']=df_liwc['je']
    df['we']=df_liwc['nous']
    df['shehe']=df_liwc['il']
    df['they']=df_liwc['ils']
    df['verb']=df_liwc['verbe']
    #affect
    df['affect']=df_liwc['affect']
    df['posemo']=df_liwc['emopos']
    df['negemo']=df_liwc['emoneg']
    #negemo
    df['anxiety']=df_liwc['anxiete']
    df['anger']=df_liwc['colere']
    df['sad']=df_liwc['tristesse']
    #attitude
    df['social']=df_liwc['social']
    df['cogproc']=df_liwc['cognition']
    df['certain']=df_liwc['certitude']
    df['past']=df_liwc['verbepasse']
    df['present']=df_liwc['verbepresent']
    df['future']=df_liwc['verbefutur']
    #other
    df['work']=df_liwc['travail']
    df['health']=df_liwc['sante']
    df['home']=df_liwc['maison']
    df['death']=df_liwc['mort'] 
    df['family']=df_liwc['famille']
    df['time']=df_liwc['temps']
    df['body']=df_liwc['corps']
    df['incl']=df_liwc['inclusion']

######################### 
def liwc_distribution(df_fr,df_en,column_name):
    one=lambda x:round(x,0)
    df_fr[column_name]=df_fr[column_name].map(one)# we reduce to only on number after the comma
    df_en[column_name]=df_en[column_name].map(one)
    df_dist_en=df_en.groupby(df_en[column_name]).count()
    df_dist_fr=df_fr.groupby(df_fr[column_name]).count()
    #print(df_dist_en.head())
    I=np.array(range(101))
    
    
    df_dist=pd.DataFrame(index=I)
    df_dist['french']=df_dist_fr['fullname']/df_dist_fr['fullname'].sum()*100
    df_dist['english']=df_dist_en['fullname']/df_dist_en['fullname'].sum()*100
    SF=[df_fr[column_name].mean(),df_fr[column_name].median(),df_fr[column_name].std()]
    SE=[df_en[column_name].mean(),df_en[column_name].median(),df_en[column_name].std()]
    
    data={'french_mean_med_std':SF,'english_mean_med_std':SE}
    df_stat=pd.DataFrame(data=data)
    df_dist_final=pd.concat([df_dist,df_stat], ignore_index=True)
    #df_dist.to_excel(column_name+'_dist'+'.xlsx')#
    return(df_dist_final)
########################comparaison
def liwc_comparaison(df_fr,df_en):# create an excel tab with 23 sheets
    writer = pd.ExcelWriter('Liwc_comparaison_2007.xlsx')
    liwc_distribution(df_fr,df_en,'work').to_excel(writer,'work')
    liwc_distribution(df_fr,df_en,'home').to_excel(writer,'home')
    liwc_distribution(df_fr,df_en,'death').to_excel(writer,'death')
    liwc_distribution(df_fr,df_en,'past').to_excel(writer,'past')
    liwc_distribution(df_fr,df_en,'future').to_excel(writer,'future')
    liwc_distribution(df_fr,df_en,'present').to_excel(writer,'present')
    liwc_distribution(df_fr,df_en,'cogproc').to_excel(writer,'cogproc')
    liwc_distribution(df_fr,df_en,'social').to_excel(writer,'social')
    liwc_distribution(df_fr,df_en,'health').to_excel(writer,'health')
    liwc_distribution(df_fr,df_en,'body').to_excel(writer,'body')
    liwc_distribution(df_fr,df_en,'family').to_excel(writer,'family')
    liwc_distribution(df_fr,df_en,'time').to_excel(writer,'time')
    liwc_distribution(df_fr,df_en,'incl').to_excel(writer,'inclusion')
    liwc_distribution(df_fr,df_en,'certain').to_excel(writer,'certain')
    liwc_distribution(df_fr,df_en,'affect').to_excel(writer,'affect')
    liwc_distribution(df_fr,df_en,'sad').to_excel(writer,'sad')
    liwc_distribution(df_fr,df_en,'anger').to_excel(writer,'anger')
    liwc_distribution(df_fr,df_en,'anxiety').to_excel(writer,'anxiety')
    liwc_distribution(df_fr,df_en,'posemo').to_excel(writer,'posemo')
    liwc_distribution(df_fr,df_en,'negemo').to_excel(writer,'negemo')
    liwc_distribution(df_fr,df_en,'cogproc').to_excel(writer,'cogproc')
    liwc_distribution(df_fr,df_en,'verb').to_excel(writer,'verb')
    liwc_distribution(df_fr,df_en,'they').to_excel(writer,'they')
    liwc_distribution(df_fr,df_en,'shehe').to_excel(writer,'shehe')
    liwc_distribution(df_fr,df_en,'we').to_excel(writer,'we')
    liwc_distribution(df_fr,df_en,'i').to_excel(writer,'i')
    liwc_distribution(df_fr,df_en,'sixltr').to_excel(writer,'sixltr')
    liwc_distribution(df_fr,df_en,'dic').to_excel(writer,'dic')
    writer.save()
#####################################
def per_category(df_dist_final):
    category_fr=[df_dist_final['french'].iloc[1:5].sum(),df_dist_final['french'].iloc[5:15].sum(),df_dist_final['french'].iloc[15:30].sum(),df_dist_final['french'].iloc[30:99].sum(),df_dist_final['french'].iloc[100],df_dist_final['french'].iloc[0]]
    category_en=[df_dist_final['french'].iloc[1:5].sum(),df_dist_final['english'].iloc[5:15].sum(),df_dist_final['english'].iloc[15:30].sum(),df_dist_final['english'].iloc[30:99].sum(),df_dist_final['english'].iloc[100],df_dist_final['english'].iloc[0]]
    index=['between 1% and 5%','between 6% and 15%','between 16% and 30%','between 31% and 99%','100%','without%']
    data={'french':category_fr,'english':category_en}
    df_cat=pd.DataFrame(index=index,data=data)
    return(df_cat)

#####################################
def liwc_comparaison_cat(df_fr,df_en):# create an excel tab with 23 sheets
    writer = pd.ExcelWriter('Liwc_comparaison_2007_cat_victim=anger.xlsx')
    per_category(liwc_distribution(df_fr,df_en,'work')).to_excel(writer,'work')
    per_category(liwc_distribution(df_fr,df_en,'home')).to_excel(writer,'home')
    per_category(liwc_distribution(df_fr,df_en,'death')).to_excel(writer,'death')
    
    per_category(liwc_distribution(df_fr,df_en,'cogproc')).to_excel(writer,'cogproc')
    per_category(liwc_distribution(df_fr,df_en,'incl')).to_excel(writer,'inclusion')
    per_category(liwc_distribution(df_fr,df_en,'certain')).to_excel(writer,'certain')
    
    per_category(liwc_distribution(df_fr,df_en,'social')).to_excel(writer,'social')
    per_category(liwc_distribution(df_fr,df_en,'family')).to_excel(writer,'family')
    
    per_category(liwc_distribution(df_fr,df_en,'health')).to_excel(writer,'health')
    per_category(liwc_distribution(df_fr,df_en,'body')).to_excel(writer,'body')
    
    
    
    per_category(liwc_distribution(df_fr,df_en,'affect')).to_excel(writer,'affect')
    per_category(liwc_distribution(df_fr,df_en,'posemo')).to_excel(writer,'posemo')
    per_category(liwc_distribution(df_fr,df_en,'negemo')).to_excel(writer,'negemo')
    per_category(liwc_distribution(df_fr,df_en,'sad')).to_excel(writer,'sad')
    per_category(liwc_distribution(df_fr,df_en,'anger')).to_excel(writer,'anger')
    per_category(liwc_distribution(df_fr,df_en,'anxiety')).to_excel(writer,'anxiety')
    
    per_category(liwc_distribution(df_fr,df_en,'time')).to_excel(writer,'time')
    
    per_category(liwc_distribution(df_fr,df_en,'verb')).to_excel(writer,'verb')
     
    per_category(liwc_distribution(df_fr,df_en,'past')).to_excel(writer,'past')
    per_category(liwc_distribution(df_fr,df_en,'future')).to_excel(writer,'future')
    per_category(liwc_distribution(df_fr,df_en,'present')).to_excel(writer,'present')
    
    per_category(liwc_distribution(df_fr,df_en,'they')).to_excel(writer,'they')
    per_category(liwc_distribution(df_fr,df_en,'shehe')).to_excel(writer,'shehe')
    per_category(liwc_distribution(df_fr,df_en,'we')).to_excel(writer,'we')
    per_category(liwc_distribution(df_fr,df_en,'i')).to_excel(writer,'i')
    
    per_category(liwc_distribution(df_fr,df_en,'sixltr')).to_excel(writer,'sixltr')
    per_category(liwc_distribution(df_fr,df_en,'dic')).to_excel(writer,'dic')
    writer.save()
```
